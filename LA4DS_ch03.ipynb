# Linear weighted combinations

```python
# the scalars
l1 = 1
l2 = 2
l3 = -3

# the vectors
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

# linear weighted combination
print(
    f'Scalar multiplication one: {l1*v1}',
    f'Scalar multiplication two: {l2*v2}',
    f'Scalar multiplication three: {l3*v3}',
    f'Linear weighted combination: {l1*v1 + l2*v2 + l3*v3}',
    sep = '\n'
)
```

# Basis vectors and points

```python
# points (in Cartesian coordinates)
p = (3,1)
q = (-6,2)

plt.figure(figsize=(8, 8))

# draw points
plt.plot(p[0],p[1],'ko',markerfacecolor='k',markersize=10,label='Point p')
plt.plot(q[0],q[1],'ks',markerfacecolor='k',markersize=10,label='Point q')

# draw basis vectors
plt.plot([0,0],[0,1],'k',linewidth=3,label='Basis S')
plt.plot([0,1],[0,0],'k',linewidth=3)

plt.plot([0,3],[0,1],'k--',linewidth=3,label='Basis T')
plt.plot([0,-3],[0,1],'k--',linewidth=3)


plt.axis('square')
plt.grid(linestyle='--',color=[.8,.8,.8])
plt.xlim([-7,7])
plt.ylim([-7,7])
plt.legend()
plt.savefig('Figure_03_04.png',dpi=300)
plt.show()
```

# Exercise 1

```python
scalars = [1, 2, -3]
vectors = [np.array([4, 5, 1]), np.array([-4, 0, -4]), np.array([1, 3, 2])]

linear_combination = np.zeros(len(vectors[0]))

for scalar, vector in zip(scalars, vectors):
    linear_combination += scalar * vector

linear_combination
```

```python
## Note: Make sure you run the code earlier to create variables l1, l2, etc.

# organized into lists
scalars = [ l1,l2,l3 ]
vectors = [ v1,v2,v3 ]

# initialize the linear combination
linCombo = np.zeros(len(v1))

# implement linear weighted combination using zip()
for s,v in zip(scalars,vectors):
  linCombo += s*v

# confirm it's the same answer as above
linCombo
```

# Exercise 2

```python
scalars = [1, 2, -3, 5, 6]
vectors = [np.array([4, 5, 1]), np.array([-4, 0, -4]), np.array([1, 3, 2]), np.array([1, 1, 1]), np.array([0, 0, 0])]

linear_combination = np.zeros(len(vectors[0]))

for scalar, vector in zip(scalars, vectors):
    linear_combination += scalar * vector

linear_combination
```

```python
# Whether it works or throws an error depends on how you set up the code.
# Using zip() as above works, because zip() will use only the minimum-matching items.
# Re-writing the code using indexing will cause an error, as in the code below.

# make the scalars longer
scalars = [ l1,l2,l3,5 ]
vectors = [ v1,v2,v3 ]

linCombo = np.zeros(len(v1))

for i in range(len(scalars)):
  linCombo += scalars[i]*vectors[i]
```

```python
# Vector set containing only one vector
vector_set = np.array([1, 3])

# 100 random numbers from uniform distribution between -4 and 4
random_scalars = np.random.uniform(-4, 4, 100)

plt.figure(figsize=(13.5, 8))

# loop over the random scalars
for scalar in random_scalars:
  plt.plot(scalar * vector_set[0], scalar * vector_set[1], 'ko', markerfacecolor='k', markersize=5)

plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid(color=[.8,.8,.8], linestyle='--')
plt.show()
```

```python
# the vector set (just one vector)
A  = np.array([ 1,3 ])

# x-axis range
xlim = [-4,4]

# random scalars in that range
scalars = np.random.uniform(low=xlim[0],high=xlim[1],size=100)



# create a figure, etc
plt.figure(figsize=(6,6))

# loop over the random scalars
for s in scalars:

  # create point p
  p = A*s

  # plot it
  plt.plot(p[0],p[1],'ko')


plt.xlim(xlim)
plt.ylim(xlim)
plt.grid()
plt.text(-4.5,4.5,'A)',fontweight='bold',fontsize=18)
plt.savefig('Figure_03_07a.png',dpi=300)
plt.show()
```

```python
# Two vectors in R3
A = np.array([3, 5, 1])
B = np.array([0, 2, 2])

# 100 random scalars from uniform distribution between -4 and 4
scalars = np.random.uniform(-4, 4, (100, 2))

# Create random points
points = np.zeros((len(scalars), len(A)))

for i in range(len(scalars)):
    points[i, :] = A * scalars[i, 0] + B * scalars[i, 1]
```

```python
import plotly

# Create 3D scatter plot
fig = plotly.graph_objects.Figure(data=[plotly.graph_objects.Scatter3d(
    x=points[:, 0],
    y=points[:, 1],
    z=points[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color=points[:, 2],                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)])

fig.update_layout(margin=dict(l=0,r=0,b=0,t=0))

# Set axes titles
fig.update_layout(scene = dict(
                    xaxis_title='x',
                    yaxis_title='y',
                    zaxis_title='z'))

# Show plot
fig.show()
```
